```python
%sql

SELECT
    course_title,
    category,
    avg_assessment_score,

    RANK() OVER(
        ORDER BY avg_assessment_score DESC
    ) AS course_rank

FROM lms_gold.gold_course_summary;
```

```python
%sql

SELECT
    instructor_name,
    avg_feedback,

    DENSE_RANK() OVER(
        ORDER BY avg_feedback DESC
    ) AS feedback_rank

FROM lms_gold.gold_instructor_summary;
```

```python
%sql

SELECT

category,

total_enrollments,

ROUND(
100.0*total_enrollments/
SUM(total_enrollments) OVER(),
2
) AS contribution_percentage

FROM lms_gold.gold_category_summary

ORDER BY contribution_percentage DESC;
```

```python
%sql

SELECT

course_title,

avg_assessment_score,

CASE

WHEN avg_assessment_score>=90 THEN 'Excellent'

WHEN avg_assessment_score>=80 THEN 'Very Good'

WHEN avg_assessment_score>=70 THEN 'Good'

WHEN avg_assessment_score>=60 THEN 'Average'

ELSE 'Poor'

END AS performance_grade

FROM lms_gold.gold_course_summary

ORDER BY avg_assessment_score DESC;
```

```python
%sql

SELECT

city,

avg_assessment_score,

avg_progress

FROM lms_gold.gold_city_summary

ORDER BY

avg_assessment_score DESC,

avg_progress DESC;
```

```python
%sql

SELECT

subscription_type,

total_learners,

avg_progress,

avg_assessment_score,

avg_feedback

FROM lms_gold.gold_subscription_summary;
```

```python
%sql

SELECT

ROUND(
100.0*
SUM(completed)/
SUM(total_enrollments),
2
)
AS completion_percentage

FROM lms_gold.gold_category_summary;
```

```python
%sql

SELECT

ROUND(AVG(avg_feedback),2)

AS overall_feedback

FROM lms_gold.gold_course_summary;
```

```python
%sql

SELECT

ROUND(AVG(avg_progress),2)

AS overall_progress

FROM lms_gold.gold_course_summary;
```

```python
%sql

SELECT

COUNT(*) AS total_courses,

ROUND(AVG(avg_progress),2) AS avg_progress,

ROUND(AVG(avg_assessment_score),2) AS avg_score,

ROUND(AVG(avg_feedback),2) AS avg_feedback,

SUM(certificates_issued) AS certificates

FROM lms_gold.gold_course_summary;
```
