# LMS Analytics Executive Dashboard

This notebook documents the creation of the LMS Analytics Executive Dashboard using Databricks SQL Dashboard.

## Objective

The objective of this dashboard is to provide business stakeholders with an executive overview of LMS performance, learner engagement, course enrollments, subscription distribution, instructor performance, and learning outcomes.

## Data Source

The dashboard is built using Gold Layer tables created during the data transformation process.

Main datasets:

- Gold Overview
- Course Summary
- Category Summary
- Instructor Summary
- City Summary
- Subscription Summary

## Dashboard Components

The dashboard contains the following visualizations:

- KPI Cards
- Top 10 Courses by Enrollment
- Enrollment Distribution by Category
- Learners by Subscription Plan
- Top 10 Instructors by Enrollment
- Top Cities by Learners
- Course Progress vs Assessment Score

## Business Insights

The dashboard helps answer questions such as:

- Which courses have the highest enrollments?
- Which instructors contribute the most enrollments?
- Which cities have the highest learner base?
- Which subscription plan is most popular?
- How does learner progress relate to assessment performance?
