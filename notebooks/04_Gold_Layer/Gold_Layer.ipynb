# Gold Layer

## Objective

The Gold Layer contains business-ready analytical tables created from the Silver Layer.

These tables are optimized for reporting, dashboards, SQL analytics, and business decision making.

```python
silver_df = spark.table("lms_silver.silver_lms")

print("Rows :", silver_df.count())
print("Columns :", len(silver_df.columns))

display(silver_df.limit(10))
```

```python
silver_df.printSchema()
```

```python
from pyspark.sql.functions import *
```

```python
silver_df = spark.table("lms_silver.silver_lms")
```

## Gold Table 1 : Overview

```python
%sql

CREATE SCHEMA IF NOT EXISTS lms_gold;
```

```python
gold_overview = silver_df.agg(

    count("*").alias("total_enrollments"),

    countDistinct("learner_id").alias("total_learners"),

    countDistinct("course_id").alias("total_courses"),

    round(avg("assessment_score"),2).alias("avg_assessment_score"),

    round(avg("feedback_rating"),2).alias("avg_feedback_rating"),

    round(avg("progress_pct"),2).alias("avg_progress")

)
```

```python
display(gold_overview)
```

```python
gold_overview.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("lms_gold.gold_overview")
```

```python
%sql

SELECT * FROM lms_gold.gold_overview;
```

## Gold Table 2 : Course Summary

```python
from pyspark.sql.functions import *
```

```python
gold_course_summary = (
    silver_df
    .groupBy(
        "course_id",
        "course_title",
        "category"
    )
    .agg(

        count("*").alias("total_enrollments"),

        round(avg("progress_pct"),2).alias("avg_progress"),

        round(avg("assessment_score"),2).alias("avg_assessment_score"),

        round(avg("feedback_rating"),2).alias("avg_feedback"),

        sum(
            when(col("status")=="Completed",1).otherwise(0)
        ).alias("completed"),

        sum(
            when(col("certificate_issued")=="Yes",1).otherwise(0)
        ).alias("certificates_issued")

    )
)
```

```python
display(gold_course_summary)
```

```python
gold_course_summary.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("lms_gold.gold_course_summary")
```

```python
%sql

SELECT *
FROM lms_gold.gold_course_summary
LIMIT 20;
```

## Gold Table 3 : Instructor Summary

```python
gold_instructor_summary = (
    silver_df
    .groupBy(
        "instructor_id",
        "instructor_name"
    )
    .agg(

        countDistinct("course_id").alias("total_courses"),

        count("*").alias("total_enrollments"),

        round(avg("assessment_score"),2).alias("avg_assessment_score"),

        round(avg("feedback_rating"),2).alias("avg_feedback"),

        round(avg("progress_pct"),2).alias("avg_progress"),

        sum(
            when(col("status")=="Completed",1).otherwise(0)
        ).alias("completed_enrollments")

    )
)
```

```python
display(gold_instructor_summary)
```

```python
gold_instructor_summary.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("lms_gold.gold_instructor_summary")
```

```python
%sql

SELECT *
FROM lms_gold.gold_instructor_summary
ORDER BY total_enrollments DESC;
```

```python
%sql

SHOW TABLES IN lms_gold;
```

## Gold Table 4 : Category Summary

```python
gold_category_summary = (
    silver_df
    .groupBy("category")
    .agg(

        count("*").alias("total_enrollments"),

        countDistinct("course_id").alias("total_courses"),

        round(avg("assessment_score"),2).alias("avg_assessment_score"),

        round(avg("feedback_rating"),2).alias("avg_feedback"),

        round(avg("progress_pct"),2).alias("avg_progress"),

        sum(
            when(col("status")=="Completed",1).otherwise(0)
        ).alias("completed")

    )
)
```

```python
display(gold_category_summary)
```

```python
gold_category_summary.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("lms_gold.gold_category_summary")
```

```python
%sql

SELECT *
FROM lms_gold.gold_category_summary
ORDER BY total_enrollments DESC;
```

```python
%sql

SHOW TABLES IN lms_gold;
```

## Gold Table 5: City Performance Summary

```python
gold_city_summary = (
    silver_df
    .groupBy("city")
    .agg(
        countDistinct("learner_id").alias("total_learners"),
        count("*").alias("total_enrollments"),
        round(avg("progress_pct"), 2).alias("avg_progress"),
        round(avg("assessment_score"), 2).alias("avg_assessment_score"),
        round(avg("feedback_rating"), 2).alias("avg_feedback"),
        sum(
            when(col("status") == "Completed", 1).otherwise(0)
        ).alias("completed_enrollments")
    )
)
```

```python
display(gold_city_summary)
```

```python
gold_city_summary.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("lms_gold.gold_city_summary")
```

```python
%sql

SELECT *
FROM lms_gold.gold_city_summary
ORDER BY total_learners DESC;
```

## Gold Table 6: Subscription Summary

```python
gold_subscription_summary = (
    silver_df
    .groupBy("subscription_type")
    .agg(
        countDistinct("learner_id").alias("total_learners"),
        count("*").alias("total_enrollments"),
        round(avg("progress_pct"), 2).alias("avg_progress"),
        round(avg("assessment_score"), 2).alias("avg_assessment_score"),
        round(avg("feedback_rating"), 2).alias("avg_feedback"),
        sum(
            when(col("status") == "Completed", 1).otherwise(0)
        ).alias("completed_enrollments")
    )
)
```

```python
display(gold_subscription_summary)
```

```python
gold_subscription_summary.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("lms_gold.gold_subscription_summary")
```

```python
%sql

SELECT *
FROM lms_gold.gold_subscription_summary
ORDER BY total_learners DESC;
```
